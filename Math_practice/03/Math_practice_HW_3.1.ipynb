```python
from sympy import *
from sympy.plotting import plot
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol("x")
```

```python
f = -3*x+5
```

```python
print(f.subs(x,5/3))
print(f.subs(x,2))
```

```python
plot(f)
```

-------------------------------

```python
f1 = 3* (x**2) + 3*x - 6
```

```python
print(f1.subs(x,0.5))
print(f1.subs(x,7))
```

```python
plot(f1)
```

-------------------------------

```python
f2 = 2* (x**4) + x**3 - 6*(x**2) + 7*x + 1
```

```python
print(f2.subs(x,3))
print(f2.subs(x,-5))
```

```python
plot(f2)
```

```python
print(f3.subs(x,0))
print(f3.subs(x,10))
```
