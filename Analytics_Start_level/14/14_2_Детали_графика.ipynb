```python
import pandas as pd
import pandas
import numpy as np
from matplotlib import pyplot as plt
data = pandas.read_csv('Tesla.csv', parse_dates=True, index_col='Date')
```

## Название графика и подпись осей

```python
plt.plot(data.index, data['High'])
plt.plot(data.index, data['Low'])
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.show()
```

## Название линий на графике и легенда

#### Автоматический режим работы функции legend

```python
plt.plot(data.index, data['High'], label='High')
plt.plot(data.index, data['Low'], label='Low')
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend()
plt.title('Tesla stock prices')
plt.show()
```

#### Явное задание соответствия линий названиям в функции legend

```python
high_prices_line, = plt.plot(data.index, data['High'])
low_prices_line, = plt.plot(data.index, data['Low'])
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend([high_prices_line, low_prices_line], ['High', 'Low'])
plt.title('Tesla stock prices')
plt.show()
```

#### Дополнительные аргументы функции legend

```python
plt.plot(data.index, data['High'], label='High')
plt.plot(data.index, data['Low'], label='Low')
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend(loc='lower right', title='Legend')
plt.title('Tesla stock prices')
plt.show()
```

## Свойства линий

Базовые свойства линий:

1. Alpha.
2. Color.
3. Marker.
4. Linestyle.
5. Linewidth.

### Использование именованных аргументов в функции plot

```python
plt.plot(data.index, data['High'], label='High', linewidth=0.5, color='g', alpha=0.5)
plt.plot(data.index, data['Low'], label='Low', linewidth=0.5, color='r', alpha=0.5)
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend(loc='lower right', title='Legend')
plt.title('Tesla stock prices')
plt.show()
```

### Использование setp для задания свойств

```python
high_prices_line, = plt.plot(data.index, data['High'], label='High', color='g')
low_prices_line, = plt.plot(data.index, data['Low'], label='Low', color='r')
plt.setp([high_prices_line, low_prices_line], linewidth=0.5, alpha=0.5)
plt.xlabel('Dates')
plt.ylabel('Stock prices')
plt.legend(loc='lower right', title='Legend')
plt.title('Tesla stock prices')
plt.show()
```

## Сетка

```python
high_prices_line, = plt.plot(data.index, data['High'], label='High', color='g')
low_prices_line, = plt.plot(data.index, data['Low'], label='Low', color='r')
plt.setp([high_prices_line, low_prices_line], linewidth=0.5, alpha=0.5)
plt.xlabel('Dates')
plt.ylabel('Stock prices, US$')
plt.grid()
plt.legend(loc='lower right', title='Legend')
plt.title('Tesla stock prices')
plt.show()
```

## Горизонтальные и вертикальные линии


```python
first_profit_report_date = pd.Timestamp(day=8, month=5, year=2013)
buy_price = 223
```

```python
high_prices_line, = plt.plot(data.index, data['High'], label='High', color='g')
low_prices_line, = plt.plot(data.index, data['Low'], label='Low', color='r')
plt.setp([high_prices_line, low_prices_line], linewidth=0.5, alpha=0.5)
plt.axvline(x=first_profit_report_date, ls='--', color='y', label=first_profit_report_date)
plt.axhline(y=buy_price, ls='--', label='Buy price {}$'.format(buy_price))
plt.xlabel('Dates')
plt.ylabel('Stock prices, US$')
plt.grid()
plt.legend(loc='lower right', title='Legend')
plt.title('Tesla stock prices')
plt.show()
```

## Размер фигуры и сохранение графика

```python
plt.figure(figsize=(18,10))
high_prices_line, = plt.plot(data.index, data['High'], label='High', color='g')
low_prices_line, = plt.plot(data.index, data['Low'], label='Low', color='r')
plt.setp([high_prices_line, low_prices_line], linewidth=0.5, alpha=0.5)
plt.axhline(y=buy_price, ls='--', label='Buy price {}$'.format(buy_price))
plt.axvline(x=first_profit_report_date, ls='--', color='y', label=first_profit_report_date)
plt.xlabel('Dates')
plt.ylabel('Stock prices, US$')
plt.grid()
plt.legend(loc='lower right', title='Legend')
plt.title('Tesla stock prices')
plt.savefig('Tesla stock prices.png')
```

## Задание штрихов на осях

```python
from pandas.tseries.offsets import MonthEnd
date = data.index[0]
dates = []
while date < data.index[-1]:
    date = date + MonthEnd(1)
    dates.append(date)
```

```python
plt.figure(figsize=(18,10))
high_prices_line, = plt.plot(data.index, data['High'], label='High', color='g')
low_prices_line, = plt.plot(data.index, data['Low'], label='Low', color='r')
plt.setp([high_prices_line, low_prices_line], linewidth=0.5, alpha=0.5)
plt.axvline(x=first_profit_report_date, ls='--', color='y', label=first_profit_report_date)
plt.axhline(y=buy_price, ls='--', label='Buy price {}$'.format(buy_price))
plt.xlabel('Dates')
plt.ylabel('Stock prices, US$')
plt.grid()
plt.xticks(dates, labels=[pd.to_datetime(str(date)).strftime("%m.%d.%Y") for date in dates], rotation='vertical')
plt.yticks(range(0, 300, 10), labels=range(0, 300, 10))
plt.legend(loc='lower right', title='Legend')
plt.title('Tesla stock prices')
plt.savefig('Tesla stock prices.png')
```
